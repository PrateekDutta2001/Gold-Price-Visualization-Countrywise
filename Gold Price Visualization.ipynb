```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
```

```python
df = pd.read_csv("annual_gold_rate.csv")
data = df.copy()
```

```python
data.head()
```

```python
data.describe().T
```

```python
data.isnull().sum()
```

```python
data["INR"].fillna(data.INR.mean(), inplace = True)
data["AED"].fillna(data.AED.mean(), inplace = True)
data["CNY"].fillna(data.CNY.mean(), inplace = True)
```

```python
data["USD"].describe().T
```
